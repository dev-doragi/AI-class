```python
X = df.drop('PRICE',axis=1) # axis=1 -> columns 제거
X.head()
```

```python
y = df['PRICE']
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

위 코드는 `sklearn` 라이브러리에서 `train_test_split` 함수를 사용하여 데이터셋을 훈련 세트와 테스트 세트로 분할하는 것을 보여줍니다. 

- `X`는 특성(feature)들로 이루어진 데이터셋이며, `y`는 해당 특성들에 대응하는 목표값(target)으로 이루어진 벡터(또는 배열)입니다.
- `test_size=0.2`는 데이터의 20%를 테스트 세트로 지정하겠다는 것을 의미합니다. 따라서 나머지 80%는 훈련 세트로 사용됩니다.
- `shuffle=True`는 데이터를 분할하기 전에 섞을 것인지 여부를 결정합니다. `random_state=12`는 섞는 과정에서 사용되는 랜덤 시드(seed)로, 이를 지정하면 항상 같은 방식으로 데이터가 섞이게 됩니다.

그 후에는 분할된 훈련 세트와 테스트 세트의 크기를 출력합니다. `X_train`은 훈련 데이터의 특성들을, `y_train`은 해당 훈련 데이터의 목표값을 나타냅니다. 마찬가지로, `X_test`와 `y_test`는 테스트 데이터의 특성과 목표값을 나타냅니다.

따라서 출력된 결과는 다음과 같습니다:
- `X_train.shape`과 `y_train.shape`는 훈련 세트의 특성과 목표값의 형태를 출력합니다.
- `X_test.shape`과 `y_test.shape`는 테스트 세트의 특성과 목표값의 형태를 출력합니다.

이런 식으로 데이터를 훈련과 테스트로 나누는 것은 모델을 훈련시키고 평가에 도움이 된다.

---

`train_test_split` 함수는 주어진 데이터를 훈련 세트와 테스트 세트로 무작위로 분할하는 데 사용됩니다. 이 함수는 주로 기계 학습 모델을 훈련하고 평가하기 위해 데이터를 준비하는 데 사용됩니다.

여기서는 `train_test_split` 함수가 받는 주요 매개변수들을 설명하겠습니다:

1. **arrays**: 분할할 데이터 배열들입니다. 이는 입력 데이터와 해당 데이터에 대응하는 목표값을 포함합니다. 보통 `X`는 입력 데이터이고 `y`는 목표값입니다. 두 배열은 데이터 포인트 간의 순서가 일치해야 합니다.

2. **test_size**: 테스트 세트의 비율을 결정하는 값입니다. 기본적으로는 0.25(25%)입니다. 예를 들어, `test_size=0.2`로 설정하면 테스트 세트가 전체 데이터셋의 20%가 됩니다.

3. **train_size**: 훈련 세트의 비율을 직접 지정할 수도 있습니다. `train_size`와 `test_size` 중 하나만 지정하거나, 둘 다 지정하지 않고 남은 부분을 자동으로 할당할 수 있습니다.

4. **random_state**: 데이터를 섞을 때 사용되는 난수 발생기의 시드(seed)입니다. 동일한 시드를 사용하면 항상 같은 방식으로 데이터가 섞입니다. 시드를 지정하지 않으면 매번 다른 방식으로 데이터가 섞입니다.

5. **shuffle**: 데이터를 분할하기 전에 섞을지 여부를 결정합니다. 기본값은 `True`입니다.

함수는 입력 데이터와 목표값에 대한 훈련 세트와 테스트 세트를 반환합니다. 일반적으로는 훈련 데이터와 테스트 데이터, 그리고 각각의 목표값에 대한 훈련 및 테스트 세트가 반환됩니다.는 데 도움이 됩니다.

## 모델 학습 후 평가하기

#### LinearRegression (선형 회귀)

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ly_preds = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(ly_preds, y_test))
print('평균제곱근오차', rmse)
```

상관관계를 보면 MedInc가 Price와 가장 관련이 높았음. \
MedInc와 Price간의 산점도를 표시하면서 실제 값과 예측 값을 비교해보자.

```python
# 예측값과 실제 값 차이를 산접도로 표시, 노란색 예측

plt.figure(figsize=(10,5))
plt.scatter(X_test['MedInc'], y_test, label='y_test')
plt.scatter(X_test['MedInc'], ly_preds, c='y', label='ly_preds')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()         
```

#### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(dy_preds, y_test))
print('평균제곱근오차', rmse)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X_test['MedInc'], y_test, label='y_test')
plt.scatter(X_test['MedInc'], dy_preds, c='g', label='dy_preds')
plt.show()         
```

#### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(ry_preds, y_test))
print('평균제곱근오차', rmse)
```
